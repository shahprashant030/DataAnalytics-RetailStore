```python
todaydate = np.datetime64('today')
TT = pd.to_datetime(df["TransactionTime"])
df['R'] = (todaydate - TT).dt.days
df.head()
```

```python
df.head(100)
```

```python
df.groupby('RetailStore')['CustomerID'].count().plot(kind='bar')
```

```python
df['F'] = df.groupby('CustomerID')['ProductID'].transform('count')
df.head()
```

```python
df['M'] = df.groupby('CustomerID')['Cost'].transform('sum')
df.head()
```

```python
df1 =df.sort_values(['R', 'F', 'M'], ascending=[True, False, False])
```

```python
df1.head()
```

```python
df1.tail()
```

```python
new_df = df1[["CustomerID","R","F","M"]].drop_duplicates()
```
